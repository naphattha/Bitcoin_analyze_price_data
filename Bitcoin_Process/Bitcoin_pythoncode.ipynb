```python
# Extracting the top correlated pairs
top_corr = correlation_matrix.unstack().sort_values(ascending=False).drop_duplicates()
print(top_corr)
```

```python
# List to store features to drop
features_to_drop = []

# Iterate through the top correlated pairs and drop one feature from each pair
for idx, corr in top_corr.items():
    feature1, feature2 = idx
    if corr > 0.95:
        # Features with a correlation coefficient greater than 0.95, drop one of them
        features_to_drop.append(feature2)

# Remove duplicates from the list of features to drop
features_to_drop = list(set(features_to_drop))

# Drop the selected features from your DataFrame
df = df.drop(features_to_drop, axis=1).copy()

# Print the list of dropped features
features_to_drop
```

```python
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```
